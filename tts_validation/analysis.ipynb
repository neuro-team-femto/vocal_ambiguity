```python
single_word = pd.read_csv('/home/paige/Documents/vocal_ambiguity/tts_validation/tts-experiment/results/french/data_exp_203859-v8/single-word.csv')
```

```python
tense = ["sheep", "pool", "cot", "peel", "fool", "rot", "bean", "wooed", "doll", "scene", "cop", "luke", "knot", "beat", "keyed", "cooed", "heat", "bought", "reap", "hot"]
lax = ["ship", "pull", "cut", "full", "pill", "rut", "bin", "wood", "dull", "sin", "cup", "look", "bit", "nut", "could", "kid", "hut", "hit", "but", "rip"]
peel = ["sheep", "peel", "bean", "scene", "beat", "keyed", "heat", "reap", "ship", "pill", "bin", "sin", "bit", "kid", "hit", "rip"]
fool = ["pool", "fool", "wooed", "luke", "cooed", "pull", "full", "wood", "look", "could"]
cut = ["cot", "rot", "doll", "cop", "knot", "bought", "cut", "rut", "dull", "cup", "nut", "hut", "hot", "but"]
```

```python
single_base = single_word[single_word['model'] == "base"]
single_emph = single_word[single_word['model'] == "emphasis"]
single_clear = single_word[single_word['model'] == "clarity"]
single_stretch = single_word[single_word['model'] == "stretch"]
```

```python
def get_wer(df):
    correct = 0
    question_count = 0
    for index, row in df.iterrows():
        if "option" in row['Response']:
            question_count = question_count + 1
            if row['word'] == row[row['Response']]:
                correct = correct + 1
    print(f"the number of correct : {correct}")
    print(question_count)
    wer = (question_count-correct)/question_count
    return wer
```

```python
def get_wer_per_type(df, model):
    tense_correct = 0
    tense_question_count = 0
    lax_correct = 0
    lax_question_count = 0
    for index, row in df.iterrows():
        if "option" in row['Response']:
            if row['word'] in tense:
                tense_question_count = tense_question_count + 1
                if row['word'] == row[row['Response']]:
                    tense_correct = tense_correct + 1
            if row['word'] in lax:
                lax_question_count = lax_question_count + 1
                if row['word'] == row[row['Response']]:
                    lax_correct = lax_correct + 1
    print(f"the number of correct : {tense_correct}")
    print(f"the number of correct : {lax_correct}")
    tense_wer = (tense_question_count-tense_correct)/tense_question_count
    lax_wer = (lax_question_count-lax_correct)/lax_question_count
    print(f"the tense error rate for {model} is : {tense_wer}")
    print(f"the lax error rate for {model} is : {lax_wer}")
```

```python
single_base_wer = get_wer(single_base)
single_emph_wer = get_wer(single_emph)
single_clear_wer = get_wer(single_clear)
single_stretch_wer = get_wer(single_stretch)

print(f"The base wer is : {single_base_wer}, the emphasis wer is : {single_emph_wer}, the stretch wer is : {single_stretch_wer}, and the clarity wer is :{single_clear_wer}]")
```

```python
get_wer_per_type(single_base, "base")
get_wer_per_type(single_emph, "emph")
get_wer_per_type(single_clear, "clear")
get_wer_per_type(single_stretch, "stretch")
```

```python
def get_wer_per_vowel(df, model):
    peel_correct = 0
    peel_question_count = 0
    fool_correct = 0
    fool_question_count = 0
    cut_correct = 0
    cut_question_count = 0
    for index, row in df.iterrows():
        if "option" in row['Response']:
            if row['word'] in peel:
                peel_question_count = peel_question_count + 1
                if row['word'] == row[row['Response']]:
                    peel_correct = peel_correct + 1
            if row['word'] in fool:
                fool_question_count = fool_question_count + 1
                if row['word'] == row[row['Response']]:
                    fool_correct = fool_correct + 1
            if row['word'] in cut:
                cut_question_count = cut_question_count + 1
                if row['word'] == row[row['Response']]:
                    cut_correct = cut_correct + 1
    print(f"the number of correct : {peel_correct}")
    print(f"the number of correct : {fool_correct}")
    print(f"the number of correct : {cut_correct}")
    peel_wer = (peel_question_count-peel_correct)/peel_question_count
    fool_wer = (fool_question_count-fool_correct)/fool_question_count
    cut_wer = (cut_question_count-cut_correct)/cut_question_count
    print(f"the error rate for words in the peel pair for {model} is : {peel_wer}")
    print(f"the error rate for words in the fool pair for {model} is : {fool_wer}")
    print(f"the error rate for words in the cut pair for {model} is : {cut_wer}")
```

```python
get_wer_per_vowel(single_base, "base")
get_wer_per_vowel(single_emph, "emph")
get_wer_per_vowel(single_clear, "clear")
get_wer_per_vowel(single_stretch, "stretch")
```

```python
def get_mos(df):
    pivoted = df.groupby('Object ID')['Response'].apply(list).to_dict()
    for key in pivoted:
        pivoted[key] = [float(item) for item in pivoted[key]]
    averages = {key: np.mean(value) for key, value in pivoted.items()}
    return averages
```

```python
def get_wer_double(df):
    correct = 0
    question_count = 0
    counter = 0
    for index, row in df.iterrows():
        if "option" in row['Response']:
            question_count = question_count + 1
            response_a, response_b = row['word'].split(',')
            if "response_a" in row['Response']:
                response, word = row['Response'].split(': ')
                if response_a == row[word]:
                    correct = correct + 1
            if "response_b" in row['Response']:
                response, word = row['Response'].split(': ')
                if response_b == row[word]:
                    correct = correct + 1
    print(f"the number of correct : {correct}")
    wer = (question_count - correct)/question_count
    return wer
```
